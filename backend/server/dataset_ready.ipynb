```python
import pandas as pd
```

```python
data1 = pd.read_csv('../dataset/coin_Bitcoin.csv')
```

```python
data1.head()
```

```python
data1['Date'] = pd.to_datetime(data1['Date']).dt.date
data1.head()
```

```python
data1.drop(['SNo', 'Name', 'Symbol'],axis=1, inplace=True)
data1.head()
```

```python
data1 = data1.head(2583)
data1.tail()
```

```python
data2 = pd.read_csv('../dataset/currency_daily_BTC_CNY.csv')
data2.head()
```

```python
data2.drop(['open (CNY)', 'high (CNY)', 'low (CNY)', 'close (CNY)'],axis=1, inplace=True)
data2.head()
```

```python
data2.rename(columns = {'timestamp':'Date', 'high (USD)': 'High', 'open (USD)': 'Open', 'low (USD)': 'Low', 'close (USD)': 'Close', 'volume': 'Volume', 'market cap (USD)': 'Marketcap'}, inplace = True)
```

```python
data2.head()
```

```python
data = pd.concat([data1, data2], ignore_index=True)
```

```python
data.to_csv('BTC.csv')
```
